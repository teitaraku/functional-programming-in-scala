```scala211
case class CreditCard(name: String)
```

```scala211
case class Charge(cc: CreditCard, amount: Double) {
    def conbine(other: Charge): Charge =
        if (cc == other.cc)
            Charge(cc, amount + other.amount)
        else
            throw new Exception("Can't combine charges to different cards")
}
```

```scala211
case class Coffee(price: Double)
```

```scala211
class Cafe {
    def buyCoffee(cc: CreditCard): (Coffee, Charge) = {
        val cup = new Coffee(5.00)
        (cup, Charge(cc, cup.price))
    }
    
    def buyCoffees(cc: CreditCard, n: Int): (List[Coffee], Charge) = {
        val purchases: List[(Coffee, Charge)] = List.fill(n)(buyCoffee(cc))
        val (coffees, charges) = purchases.unzip
          (coffees, charges.reduce((c1, c2) => c1.conbine(c2)))
    }
}
```

```scala211
val cafe = new Cafe()
//cafe.buyCoffee(CreditCard("Tom"): CreditCard)
val cc = CreditCard("Tom")
cafe.buyCoffee(cc)
```
